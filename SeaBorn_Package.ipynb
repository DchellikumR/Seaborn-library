## Correlation with heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale to show in 2d correlation matrix. It is very importatnt in feature engineering.
Correlation value will be ranging from -1 to +1.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## joint plot
Joint plot helps us to do univarient analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

### PAIR PLOT                                                        
A pair plot is also know as scatterplot, In which one variable in the same data row is matched with another variable value.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['smoker'].value_counts()
```

# DIST Plot
Dist plot helps us to check the distribution of the column feature

```python
# here it won't show the count value on y axis because KDE(Kernal density estimation) is true by default
sns.distplot(df['tip'])
```

```python
# KDE is false
sns.distplot(df['tip'], kde=False)
```
